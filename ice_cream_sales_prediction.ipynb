#Ice-Cream Sales Predictions

```python
#step 1 -> import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#step 2 -> Creating Dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Ice Cream.csv')
df = pd.DataFrame(data)
df
```

```python
#step 3 -> taking randomly 30 samples
df_30 = df.sample(n=30).reset_index(drop = True)
df_30
```

```python
#Step 4 -> Data Visualization
plt.xlabel('Temperature', color = 'red')
plt.ylabel('Revenue', color = 'green')
plt.title('Ice-craem Sales Prediction based on Temperature', color = 'purple')
plt.scatter(df.Temperature, df.Revenue)
```

```python
#Step 5 -> Model Training
lreg = LinearRegression()
lreg.fit(df_30[['Temperature']], df_30['Revenue']) #Here Temperature is feature and Revenue is our Target
```

```python
#Step 6 -> Predictions
y_predicted = lreg.predict([[50]])
print("Predicted value: ", y_predicted)

#Step 7 -> Chceking Accuracy
Accuracy = lreg.score(df_30[['Temperature']], df_30.Revenue)
print("Accuracy is: ", Accuracy)
```

```python
#Step # 8 -> checking coefficient & intercept
coefficient = lreg.coef_
print("Coefficient is: ", coefficient)
intercept = lreg.intercept_
print("Intercept is: ", intercept)

predicted = coefficient * 50 + intercept
print("Predicted value: ", predicted)
```

```python
#Step 9 -> Linear
plt.xlabel('Temperature', color = 'red')
plt.ylabel('Revenue', color = 'green')
plt.title('Ice-craem Sales Prediction based on Temperature', color = 'purple')
plt.scatter(df_30.Temperature, df_30.Revenue)
plt.plot(df_30.Temperature, lreg.predict(df_30[['Temperature']]), color = 'red')
```
